```python
#Prediction with trained ann model
import tensorflow as tf
from tensorflow.keras.models import load_model
import pickle
import pandas as pd
import numpy as np
```

```python
# load the trained model ,sclaer, pickle, onehot 

model = load_model('ann_model.h5')

# load encoder and scaler
with open('onehot_encoder_geo.pk1','rb') as f:
    label_encoder_geo = pickle.load(f)
with open('label_encoder_gender.pk1','rb') as f:
    label_encoder_gender = pickle.load(f)
with open('scaler.pk1', 'rb') as f:
    scaler = pickle.load(f)
```

```python
# loading data

input_data= {
    'CreditScore':50000,
    'Geography':'France',
    'Gender':'Female',
    'Age':40,
    'Tenure':4,
    'Balance':60000,
    'NumOfProducts':3,
    'HasCrCard':1,
    'IsActiveMember':1,
    'EstimatedSalary':50000
}
```

```python
# onehot encoding 'Geography'

geo_encoded = label_encoder_geo.transform([[input_data['Geography']]]).toarray()
geo_encoded_df = pd.DataFrame(geo_encoded, columns =label_encoder_geo.get_feature_names_out(['Geography']))
geo_encoded_df


```

```python
# converted dictionary data into dataframe

inputdf = pd.DataFrame([input_data])
inputdf
```

```python
inputdf['Gender']= label_encoder_gender.transform(inputdf['Gender'])

inputdf
```

```python
# concatination of onehot encoded 
input_df=pd.concat([inputdf.drop("Geography", axis=1), geo_encoded_df], axis=1)
input_df
```

```python
#Scaling the input data
input_scaled = scaler.transform(input_df)
input_scaled
```

```python
#predicting the churn

prediction = model.predict(input_scaled)
prediction
```

```python
# probaility of the prediction
prediction_prob= prediction[0][0]
prediction_prob
```

```python
if prediction_prob>0.5:
    print("The customer is likely to churn.")
else:
    print("The customer is not likely to churn")
```
